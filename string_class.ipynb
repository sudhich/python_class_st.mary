# Padding function
### 1=> rjust()
### 2=> ljust()
### 3=> center()
### 4=> zfill()


```python
var="Pyhton1"
print(var)
print(var.rjust(10))
print(var.ljust(10))
print(var.center(10))
print(var.zfill(10))
```

```python
var="Pyhton"
print(var)
print(var.rjust(10,'-'))
print(var.ljust(10,"@"))
print(var.center(11,"$"))
print(var.zfill(10))
```

# String Substitution Function
### 1=> replace(old_value,new_value,count)
### 2=> translate(table)
### 3=> split(pattern,count)
### 4=> splitlines()
### 5=> join()

```python
txt = "one one was a race horse, two two was one too."
print(txt)
x = txt.replace("one", "three")
print(x)
y = txt.replace("one", "three",2)
print(y)
```

```python
txt = "one one was a race horse, two two was one too."
print(txt.replace("one", "three", 2))
```

```python
print("one one was a race horse, two two was one too.".replace("one", "three", 2))
```

```python
str7="Hello, World"
print(str7)
table={ord("o"):"L",ord(","):None}
print(str7.translate(table))
```
